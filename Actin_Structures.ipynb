```python
#Import stuff
%matplotlib inline
from bokeh.plotting import figure, show, output_notebook
# from bokeh.charts import Scatter, output_file, show
from bkcharts import Scatter, output_file, show
from bokeh.palettes import Spectral4
from bokeh.layouts import gridplot
from bokeh.models import HoverTool

import matplotlib
import numpy as np 
import matplotlib.pyplot as plt
```

### Overview

```python
factin_dir = "/home/kmm5/actin/sbm/pdb_maps_8A/all_pdb_monomer_8A.contacts"
factin_inter_dir = "/home/kmm5/actin/sbm/pdb_maps_8A/all_pdb_interface_8A.contacts"
gactin_dir = "/home/kmm5/actin/gactin/8A/cutoff/gactin_ca_8A.contacts"
dca_dir = "/home/kmm5/actin/plmDCA/pfam/pfam31/mapped_act_pfam31_f10.contacts"
# dca_dir = "plmDCA/cytoplasm/from_actin-actin_subseq/actinx2_09-25-17/actx2_ranked.fn"
```

```python
Fi, Fj = np.loadtxt(factin_dir, unpack=True)
Ii, Ij = np.loadtxt(factin_inter_dir, unpack=True)
Gi, Gj = np.loadtxt(gactin_dir, unpack=True)
pi, pj = np.loadtxt(dca_dir, usecols=(0,1), unpack=True)
```

```python
N = 200
hover = HoverTool(tooltips=[('$Res i, $Res j', '$x{0}, $y{0}')])
TOOLS = "crosshair, pan, wheel_zoom, reset, save, box_select"
p = figure(title=('Top %d DCA' % N), 
           width=705, plot_height=500, tools=[TOOLS, hover])
p.square(Fi, Fj, 
         legend='F-actin Monomer', alpha=0.5, color='navy')
p.square(Ii, Ij, 
         legend='F-actin Interface', alpha=0.5, color='green')
# p.scatter(Gi, Gj, 
#           legend='G-actin Monomer', alpha=0.5, color='green')
p.scatter(pi[:N], pj[:N], 
          legend='DCA', alpha=0.8, color='orangered', 
          line_width=1, size=4)
p.legend.location="bottom_right"
p.legend.click_policy="hide"
output_notebook()
show(p)
```
